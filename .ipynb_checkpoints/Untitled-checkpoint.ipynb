```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 2.5, 3, 4]
y = [1, 4, 7, 9, 15]

plt.plot(x,y, 'ro')
# plt.plot(x,y, 'g') #With no parameters it will just make a line connecting all of them
plt.axis([0,7,0,20])
# print((np.unique(x)))#Returns all Unique values in the array(x)
# print(np.polyfit(x, y, 1))#Gets actual values for y = mx + b
# print(np.poly1d(np.polyfit(x, y, 1))) # Returns the formula for given values (m)
# print(np.poly1d(np.polyfit(x, y, 1))(np.unique(x))) # Returns the Array of the (y = mx)
plt.plot(np.unique(x), np.poly1d( np.polyfit(x, y, 1) ) ( np.unique(x) ) )   # y = m x + b
plt.show()
```
